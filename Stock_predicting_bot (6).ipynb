```python
# Imports
import pandas as pd
import numpy as np
from pathlib import Path
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
# Read the 4 csv files into Pandas DataFrames
Nasdaq_2010_df = pd.read_csv(
   Path("NASDAQ100_LIST_2010.csv"))

Nasdaq_2011_df = pd.read_csv(
   Path("NASDAQ100_LIST_2011.csv"))

Nasdaq_2012_df = pd.read_csv(
   Path("NASDAQ100_LIST_2012.csv"))

Ten_x_df = pd.read_csv(
   Path("NASDAQ100_y_Values.csv"))

# Review the DataFrames
Nasdaq_2010_df
```

```python
Nasdaq_2010_df = Nasdaq_2010_df.drop(columns=['Price 2020'])
Nasdaq_2010_df.columns = ['TICKER','2010 P/E', '2010 Sales Surprise', '2010 P/B', '2010 D/E', '2010 ROE','2010 FCF','2010 PEG','2010 Market cap','2010 EBITDA','2010 beta','2010 Price']
Nasdaq_2010_df = Nasdaq_2010_df.iloc[1: , :]

# Review the DataFrame
Nasdaq_2010_df
```

```python
Nasdaq_2011_df = Nasdaq_2011_df.drop(columns=['Price 2021'])
Nasdaq_2011_df.columns = ['TICKER','2011 P/E', '2011 Sales Surprise', '2011 P/B', '2011 D/E', '2011 ROE','2011 FCF','2011 PEG','2011 Market cap','2011 EBITDA','2011 beta','2011 Price']
Nasdaq_2011_df = Nasdaq_2011_df.iloc[1: , :]

# Review the DataFrame
Nasdaq_2011_df
```

```python
Nasdaq_2012_df = Nasdaq_2012_df.drop(columns=['Price 2022'])
Nasdaq_2012_df.columns = ['TICKER','2012 P/E', '2012 Sales Surprise', '2012 P/B', '2012 D/E', '2012 ROE','2012 FCF','2012 PEG','2012 Market cap','2012 EBITDA','2012 beta','2012 Price']
Nasdaq_2012_df = Nasdaq_2012_df.iloc[1: , :]

# Review the DataFrame
Nasdaq_2012_df
```

```python
Ten_x_df = Ten_x_df.iloc[1: , :]
Ten_x_df = Ten_x_df.drop(columns=['Price 2012','price max'])
Ten_x_df.columns = ['TICKER', 'Y']
Ten_x_df = Ten_x_df.set_index('TICKER')

# Review the DataFrame
Ten_x_df
```

```python
# Merge 4 Dataframes
Nasdaq_df = pd.merge(pd.merge(pd.merge(Nasdaq_2010_df,Nasdaq_2011_df,on='TICKER'),Nasdaq_2012_df,on='TICKER'),Ten_x_df,on='TICKER')

# Set index
#Nasdaq_df = Nasdaq_df.set_index('TICKER')

# Drop dupes
Nasdaq_df = Nasdaq_df.drop_duplicates()

# Change #LinkErrors to NAs
Nasdaq_df = Nasdaq_df.replace('#LinkError', np.nan)

# Drop NAs
Nasdaq_df = Nasdaq_df.dropna()

# Review the DataFrame
#type(Nasdaq_df)

#print(Nasdaq_df.dtypes)
```

```python
# Define the target set y
y = Nasdaq_df['Y']

# Display y
y
```

```python
# Define features X
X = Nasdaq_df.drop(columns=['TICKER','Y'])

# Review the features DataFrame
X = X.astype('float')
print(X.dtypes)
X
```

```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to the features training dataset
X_scaler = scaler.fit(X_train)

# Fit the scaler to the features training dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
# Define the the number of inputs (features) to the model
number_input_features = len(X_train.iloc[0])

# Review the number of features
number_input_features
```

```python
# Define the number of neurons in the output layer
number_output_neurons = 1
```

```python
# Define the number of hidden nodes for the first hidden layer
hidden_nodes_layer1 = 99# (number_input_features + 1) // 2 

# Review the number hidden nodes in the first layer
hidden_nodes_layer1
```

```python
# Define the number of hidden nodes for the second hidden layer
hidden_nodes_layer2 = (hidden_nodes_layer1 + 1) // 2

# Review the number hidden nodes in the second layer
hidden_nodes_layer2
```

```python
# Create the Sequential model instance
nn = Sequential()
```

```python
# Add the first hidden layer
nn.add(Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))
```

```python
# Add the second hidden layer
nn.add(Dense(units=hidden_nodes_layer2, activation="relu"))
```

```python
# Add the output layer to the model specifying the number of output neurons and activation function
nn.add(Dense(units=1, activation="sigmoid"))
```

```python
# Display the Sequential model summary
nn.summary()
```

```python
# Compile the Sequential model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model using 50 epochs and the training data
fit_model = nn.fit(X_train, y_train, epochs=50)
```

```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
